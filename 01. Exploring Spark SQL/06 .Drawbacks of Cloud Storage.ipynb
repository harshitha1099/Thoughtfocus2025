# Prerequisites

## Load users_001.csv into DataFrame 

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
df.display()
 
```

## Write into the Parquet Format

```python
df.write.format("parquet").save(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet",
    mode="overwrite",
)
spark.read.parquet(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet"
).display()
```

# Create a View

```python
df_parquet= spark.read.parquet("/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet")
```

```python
df_parquet.createOrReplaceTempView("users_vw")
```

## Disadvantage -  Update is not allowed

```python
%sql
 
UPDATE users_vw
SET country = 'Bharat'
where country = 'India'
```
