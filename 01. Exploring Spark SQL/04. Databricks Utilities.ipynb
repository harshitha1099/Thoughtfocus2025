```python
dbutils.widgets.combobox("top_n_records",
                         defaultValue="1",
                         choices=["1","2","3","4"],
                         label="Top N Records")
```

```python
dbutils.widgets.get("top_n_records")
```

```python
#Usage of Dropdown

dbutils.widgets.dropdown(name = "EnvironmentType",
                         defaultValue="Development",
                         choices=["Development", "Testing", "Pre Production"],
                         label="Environment")
```

```python
# multiselect
dbutils.widgets.multiselect(name = "user_input",
                         defaultValue="2G",
                         choices=["2G", "3G", "4G"],
                         label="Technology")
```

```python
#to add or remove widgets

dbutils.widgets.remove("top_n_records")
```

```python
dbutils.widgets.removeAll() #remove all widgets
```
